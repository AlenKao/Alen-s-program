```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
y = (25-(x**2))**0.5
z = -(25-(x**2))**0.5
plt.plot(x, y)
plt.plot(x, z)
```

```python
x1 = np.linspace(0, 5, 100)
y1 = -1.2*x + 3
plt.plot(x1, y1)

x2 = np.linspace(-5, 0, 100)
y2 = 1.2*x + 3
plt.plot(x2, y2)

plt.plot([-5, 5], [-3, -3])
```

```python
plt.plot([-5, 5], [3, 3])
plt.plot([-5, 5], [-3, -3])
plt.plot([-5, -5], [3, -3])
plt.plot([5, 5], [3, -3])
```
